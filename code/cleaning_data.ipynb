```python
df = pd.read_csv('../csv/1900_2021_disasters.csv')
df = df[['Year', 'Disaster Type', 'Disaster Subtype', 'Event Name', 'Country', 'Location']]
df.columns = ['Year', 'Type', 'Subtype', 'Name', 'Country', 'Location']

df.head()
```

```python
disaster_type = np.sort(df['Type'].unique())

for item in disaster_type:
    label = item.replace(' ', '_')
    fileName = '../csv/' + label + '.csv'
    df_tmp = df[df['Type'] == item]
    df_tmp.sort_values(by=['Country', 'Subtype', 'Year'], inplace=True)
    df_tmp.to_csv(fileName, index=False)
```
